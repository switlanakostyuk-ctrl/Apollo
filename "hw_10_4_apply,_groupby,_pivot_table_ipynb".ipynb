–í —Ü—å–æ–º—É –¥–æ–º–∞—à–Ω—å–æ–º—É –∑–∞–≤–¥–∞–Ω–Ω—ñ –ø—Ä–æ–¥–æ–≤–∂—É—î–º–æ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ –Ω–∞–±–æ—Ä–æ–º –¥–∞–Ω–∏—Ö `supermarket_sales.csv`.

0. –Ü–º–ø–æ—Ä—Ç—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É pandas —Ç–∞ –∑—á–∏—Ç–∞–π—Ç–µ –¥–∞–Ω—ñ —É –∑–º—ñ–Ω–Ω—É `df` —Ç–∏–ø—É `pandas.DataFrame`.

```python
import pandas as pd
df = pd.read_csv('supermarket_sales.csv')
path = '/content/drive/MyDrive/supermarket_sales.csv'
```

1. –î–æ—Å–ª—ñ–¥–∏–º–æ, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª —Å—É–ø–µ—Ä–º–∞—Ä–∫–µ—Ç—É ('Branch') —î –Ω–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à–∏–º. –î–ª—è —Ü—å–æ–≥–æ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑–∞ –∫–æ–∂–Ω–∏–º —Ñ—ñ–ª—ñ–∞–ª–æ–º —ñ –≤–∏—è–≤—ñ—Ç—å, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª –º–∞—î –Ω–∞–π–≤–∏—â–∏–π.

```python
profit_by_branch = df.groupby("Branch")["gross income"].sum()
profit_by_branch
```

```python
best_branch = profit_by_branch.idxmax()
best_branch
```

2. –í —è–∫–æ–º—É –º—ñ—Å—Ç—ñ –∑–Ω–∞–π—Ö–æ–¥–∏—Ç—å—Å—è —Ñ—ñ–ª—ñ–∞–ª –∑ –Ω–∞–π–≤–∏—â–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º? –ú–æ–∂–µ –≤ —Ç–æ–º—É –º—ñ—Å—Ü—ñ –Ω–∞–º —Ä–æ–∑–º—ñ—Ç–∏—Å—Ç–∏ —â–µ –æ–¥–∏–Ω –º–∞–≥–∞–∑–∏–Ω.  
–ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—å –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É Pandas.

```python
city_of_best_branch = df[df["Branch"] == best_branch]["City"].iloc[0]
city_of_best_branch
```

3.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –∑–≤–æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—é, —è–∫–∞ –ø–æ–∫–∞–∂–µ, —Å–∫—ñ–ª—å–∫–∏ –ø–æ–∫—É–ø–æ–∫ (—ñ–Ω–≤–æ–π—Å—ñ–≤) –±—É–ª–æ –∑—Ä–æ–±–ª–µ–Ω–æ –≤ –∫–æ–∂–Ω—ñ–π –∑ —Ñ—ñ–ª—ñ–π (`Branch`) –∑–∞ —Ä—ñ–∑–Ω–∏–º–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è–º–∏ —Ç–æ–≤–∞—Ä—ñ–≤. –ó–∞–ø–∏—à—ñ—Ç—å —Ç–∞–±–ª–∏—Ü—é –≤ –∑–º—ñ–Ω–Ω—É `invoices_by_category` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –∑–º—ñ–Ω–Ω—É –Ω–∞ –µ–∫—Ä–∞–Ω.
–¶—è —Ç–∞–±–ª–∏—Ü—è –¥–æ–ø–æ–º–æ–∂–µ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –≤ —è–∫—ñ–π —Ñ—ñ–ª—ñ—ó –∫—É–ø—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∫–æ–∂–Ω–æ—ó –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.

```python
invoices_by_category = pd.pivot_table(
    df,
    values="Invoice ID",
    index="Branch",
    columns="Product line",
    aggfunc="count"
)

invoices_by_category
```

3.2. –í–∏–∫—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –∑–º—ñ–Ω–Ω—É `invoices_by_category` –¥–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –ø—Ä–æ–≥—Ä–∞–º–Ω–æ (—Ç–æ–±—Ç–æ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç—Ä–µ–±–∞ –Ω–µ –ø—Ä–æ—Å—Ç–æ –∑–Ω–∞–π—Ç–∏ –æ—á–∏–º–∞, –∞ –≤–∏–≤–µ—Å—Ç–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∫–æ–¥—É), –≤ —è–∫–æ–º—É —Ñ—ñ–ª—ñ–∞–ª—ñ –º–∞–≥–∞–∑–∏–Ω—É (`Branch`) –Ω–∞–π–±—ñ–ª—å—à–µ —ñ–Ω–≤–æ–π—Å—ñ–≤ —ñ–∑ –ø–æ–∫—É–ø–∫–æ—é —Ç–æ–≤–∞—Ä—ñ–≤ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó "–ï–ª–µ–∫—Ç—Ä–æ–Ω–Ω—ñ –∞–∫—Å–µ—Å—É–∞—Ä–∏" (`Electronic accessories`)?


```python
best_branch_electronics = invoices_by_category["Electronic accessories"].idxmax()
best_branch_electronics
```

4-6. **–¢–≤–æ—Ä—á–µ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ —Ä–æ–∑–≤–∏—Ç–æ–∫ –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–æ–≥–æ –º–∏—Å–ª–µ–Ω–Ω—è**

–ö—Ä–æ–∫ 1. –°—Ñ–æ—Ä–º—É–ª—é–π—Ç–µ –¢–†–ò –ø–∏—Ç–∞–Ω–Ω—è (–≥—ñ–ø–æ—Ç–µ–∑–∏) –¥–æ –Ω–∞—è–≤–Ω–∏—Ö –¥–∞–Ω–∏—Ö, —è–∫—ñ –¥–æ–ø–æ–º–æ–≥–ª–∏ –± –≤–∞–º –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ —â–æ, –¥–µ —Ç–∞ –∫–æ–ª–∏ –Ω–∞–π–±—ñ–ª—å—à–µ/–Ω–∞–π–º–µ–Ω—à–µ –∫—É–ø—É—é—Ç—å, –∞–±–∏ –¥–∞—Ç–∏ –Ω–∞ –æ—Å–Ω–æ–≤—ñ —Ü–∏—Ö –≥—ñ–ø–æ—Ç–µ–∑ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—ó –±—ñ–∑–Ω–µ—Å—É. –ó–≤—ñ—Å–Ω–æ –ø–∏—Ç–∞–Ω–Ω—è –º–∞—é—Ç—å –±—É—Ç–∏ –Ω–µ —Ç–∏–º–∏, –Ω–∞ —è–∫—ñ –º–∏ –≤–∂–µ –≤—ñ–¥–ø–æ–≤—ñ–ª–∏ –≤ –∑–∞–≤–¥–∞–Ω–Ω—è—Ö –º–æ–¥—É–ª—é.

–ö—Ä–æ–∫ 2. –ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ –Ω–∞ —Å–≤–æ—ó –ø–∏—Ç–∞–Ω–Ω—è –∑ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É pandas.

–ö—Ä–æ–∫ 3. –ù–∞–ø–∏—à—ñ—Ç—å, —è–∫ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ —Ü–µ –ø–∏—Ç–∞–Ω–Ω—è –º–æ–∂–µ –±—É—Ç–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–∞ –¥–ª—è –ø—Ä–∏–π–Ω—è—Ç—Ç—è –±—ñ–∑–Ω–µ—Å —Ä—ñ—à–µ–Ω—å.   
   
 –ü–∏—Ç–∞–Ω–Ω—è –º–æ–∂—É—Ç—å –±—É—Ç–∏ –±—É–¥—å-—è–∫–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ, –∞–ª–µ —ó—Ö –º–∞—î –±—É—Ç–∏ 3. –ö–æ–∂–Ω–µ –ø–∏—Ç–∞–Ω–Ω—è –æ—Ü—ñ–Ω—é—î—Ç—å—Å—è —è–∫ 1 –∑–∞–≤–¥–∞–Ω–Ω—è. –ë–µ–∑ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –î–ó –Ω–µ –ø—Ä–∏–π–º–∞—î—Ç—å—Å—è. –Ø–∫—â–æ —î –ø–∏—Ç–∞–Ω–Ω—è —â–æ–¥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è - –ø–∏—à—ñ—Ç—å —É —á–∞—Ç üôå


–ì—ñ–ø–æ—Ç–µ–∑–∞ 1. –í–ø–ª–∏–≤ —Å—Ç–∞—Ç—ñ –Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤.

–Ø–∫–∞ —Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ –≤–∏—Ç—Ä–∞—Ç–∞–º–∏ –Ω–∞ –∂—ñ–Ω–æ—á—ñ —ñ —á–æ–ª–æ–≤—ñ—á—ñ —Ç–æ–≤–∞—Ä–∏?

```python
gender_preferences = df.pivot_table(index='Product line', columns='Gender', values='Total', aggfunc='mean')
print(gender_preferences)
```

–ë—ñ–∑–Ω–µ—Å-—Ä—ñ—à–µ–Ω–Ω—è: —á–æ–ª–æ–≤—ñ–∫–∏ –≤–∏—Ç—Ä–∞—á–∞—é—Ç—å –±—ñ–ª—å—à–µ –≤ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó "Health and beauty", –∞ –∂—ñ–Ω–∫–∏ –≤ "Electronic accessories", –º–∏ –º–æ–∂–µ–º–æ –∑–º—ñ–Ω–∏—Ç–∏ —Ç–∞—Ä–≥–µ—Ç —Ä–µ–∫–ª–∞–º–∏ –≤ —Å–æ—Ü–º–µ—Ä–µ–∂–∞—Ö –∞–±–æ –ø–µ—Ä–µ—Å—Ç–∞–≤–∏—Ç–∏ –ø–æ–ª–∏—Ü—ñ –≤ –º–∞–≥–∞–∑–∏–Ω—ñ —Ç–∞–∫, —â–æ–± –ø–æ–ø—É–ª—è—Ä–Ω—ñ —Ç–æ–≤–∞—Ä–∏ –±—É–ª–∏ –Ω–∞ –≤–∏–¥—É —É –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ—ó –∞—É–¥–∏—Ç–æ—Ä—ñ—ó.

–ì—ñ–ø–æ—Ç–µ–∑–∞ 2. –ï—Ñ–µ–∫—Ç–∏–≤–Ω—ñ—Å—Ç—å –ø–æ–∫—É–ø–æ–∫ –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –¥–Ω—è —Ç–∏–∂–Ω—è (Weekday Analysis)
–£ —è–∫—ñ –¥–Ω—ñ —Ç–∏–∂–Ω—è —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä–µ–π—Ç–∏–Ω–≥ –∞–±–æ —Å—É–º–∞ —á–µ–∫–∞ –ø–∞–¥–∞—é—Ç—å?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.day_name()
```

```python
weekday_analysis = df.groupby('Weekday')[['Total', 'Rating']].mean().sort_values(by='Total', ascending=False)
print(weekday_analysis)
```

 –°—É–±–æ—Ç–∞: –ù–∞–π–≤–∏—â–∏–π —á–µ–∫ ($342$), –∞–ª–µ –æ–¥–∏–Ω –∑ –Ω–∞–π–Ω–∏–∂—á–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥—ñ–≤ ($6.9$).–ü–æ–Ω–µ–¥—ñ–ª–æ–∫: –ù–∞–π–Ω–∏–∂—á–∏–π —á–µ–∫ ($303$), –∞–ª–µ –Ω–∞–π–≤–∏—â–∏–π —Ä–µ–π—Ç–∏–Ω–≥ ($7.15$)

–†—ñ—à–µ–Ω–Ω—è: –£ —Å—É–±–æ—Ç—É —Ç–∞ –Ω–µ–¥—ñ–ª—é –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–≤–æ–¥–∏—Ç–∏ –±—ñ–ª—å—à–µ –ø–µ—Ä—Å–æ–Ω–∞–ª—É –Ω–∞ –∑–º—ñ–Ω—É (–∫–∞—Å–∏—Ä—ñ–≤, –∫–æ–Ω—Å—É–ª—å—Ç–∞–Ω—Ç—ñ–≤), —â–æ–± –∑–º–µ–Ω—à–∏—Ç–∏ —á–µ—Ä–≥–∏ —Ç–∞ –ø—ñ–¥–≤–∏—â–∏—Ç–∏ —è–∫—ñ—Å—Ç—å —Å–µ—Ä–≤—ñ—Å—É. –£ –ø–æ–Ω–µ–¥—ñ–ª–æ–∫, –∫–æ–ª–∏ –∫–ª—ñ—î–Ω—Ç—ñ–≤ –º–∞–ª–æ, –ø–µ—Ä—Å–æ–Ω–∞–ª –ø—Ä–∞—Ü—é—î –∫—Ä–∞—â–µ ‚Äî –º–æ–∂–Ω–∞ –ø–æ–¥—É–º–∞—Ç–∏ –ø—Ä–æ —Å–∫–æ—Ä–æ—á–µ–Ω–Ω—è –∑–º—ñ–Ω–∏ –∞–±–æ –ø—Ä–æ–≤–µ–¥–µ–Ω–Ω—è –Ω–∞–≤—á–∞–Ω–Ω—è —Å–∞–º–µ –≤ —Ü–µ–π —á–∞—Å.

–ì—ñ–ø–æ—Ç–µ–∑–∞ 3. –ï—Ñ–µ–∫—Ç–∏–≤–Ω—ñ—Å—Ç—å –ø–æ–∫—É–ø–æ–∫ —É –≤–µ—á—ñ—Ä–Ω—ñ–π —á–∞—Å (Happy Hours)
–ß–∏ –∑–º—ñ–Ω—é—î—Ç—å—Å—è —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä–µ–π—Ç–∏–Ω–≥ –ø—ñ—Å–ª—è 18:00? –í–µ—á—ñ—Ä ‚Äî —Ü–µ –ø—ñ–∫ –ø–æ–∫—É–ø–æ–∫, –∞–ª–µ —á–∏ –∑–∞–ª–∏—à–∞—é—Ç—å—Å—è –∫–ª—ñ—î–Ω—Ç–∏ –∑–∞–¥–æ–≤–æ–ª–µ–Ω–∏–º–∏?

```python
def get_hour(time_str):
    return int(time_str.split(':')[0])

df['Hour'] = df['Time'].apply(get_hour)
```

```python
df['Is Evening'] = df['Hour'] >= 18
```

```python
evening_efficiency = df.groupby('Is Evening')[['Rating', 'Total']].mean()
print(evening_efficiency)
```
