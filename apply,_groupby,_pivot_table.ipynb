3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = pd.pivot_table(
    df,
    values="Invoice ID",
    index="Branch",
    columns="Product line",
    aggfunc="count"
)

invoices_by_category
```

Гіпотеза 2. Ефективність покупок залежно від дня тижня (Weekday Analysis)
У які дні тижня середній рейтинг або сума чека падають?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.day_name()
```

```python
weekday_analysis = df.groupby('Weekday')[['Total', 'Rating']].mean().sort_values(by='Total', ascending=False)
print(weekday_analysis)
```

 Субота: Найвищий чек ($342$), але один з найнижчих рейтингів ($6.9$).Понеділок: Найнижчий чек ($303$), але найвищий рейтинг ($7.15$)

Дані показують зворотну кореляцію: коли грошей багато (вихідні), задоволеність клієнтів падає.

Рішення: У суботу та неділю потрібно виводити більше персоналу на зміну (касирів, консультантів), щоб зменшити черги та підвищити якість сервісу. У понеділок, коли клієнтів мало, персонал працює краще — можна подумати про скорочення зміни або проведення навчання саме в цей час.

Гіпотеза 3. Ефективність покупок у вечірній час (Happy Hours)
Чи змінюється середній рейтинг після 18:00? Вечір — це пік покупок, але чи залишаються клієнти задоволеними?

```python
def get_hour(time_str):
    return int(time_str.split(':')[0])

df['Hour'] = df['Time'].apply(get_hour)
```

```python
df['Is Evening'] = df['Hour'] >= 18
```

```python
evening_efficiency = df.groupby('Is Evening')[['Rating', 'Total']].mean()
print(evening_efficiency)
```

Вдень: Рейтинг 6.98, середній чек $325.82

Ввечері: Рейтинг 6.94, середній чек $315.65

Бізнес-рішення: середній чек ввечері нижчий, і клієнти менш задоволені (рейтинг падає).

Чому так? Ймовірно, ввечері люди забігають за дрібницями після роботи, а великі черги псують настрій.
