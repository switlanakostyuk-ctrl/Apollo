# Найприбутковіші чеки

```python
# 1. Значення 75-го персентиля
q75 = df['gross income'].quantile(0.75)
```

```python
# 2. Записи, що вище або дорівнюють q75
top_25_sales = df[df['gross income'] >= q75]
```

```python
# 3. Результат
print(f"Кількість записів у топових 25%: {top_25_sales.shape[0]}")
top_25_sales.head()
```
