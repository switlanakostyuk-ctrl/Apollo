4. Напишіть функцію `multipy_numbers`, яка приймає два числа як аргументи і повертає їх добуток.

  Викличіть функцію `multipy_numbers` на парах чисел
- 10 і 2
- 7 і 8
- 1024 і 999.
  
  Виведіть результат кожного виклику.
  При виконанні цього завдання з використанням циклу - отримаєте 2х балів за нього :)

````python
def multipy_numbers(a, b):
    return a * b

pairs = [(10, 2), (7, 8), (1024, 999)]

for a, b in pairs:
    print(multipy_numbers(a, b))
````

5. **Ділення з перевіркою типів.**  

  1. Напишіть функцію `int_division` яка приймає на вхід два аргументи типу `int` - `a` і `b`, і виконує ділення першого аргумента на другий: `a/b`. При цьому функція перед діленням перевіряє, чи другий аргумент не дорівнює 0. Якщо другий аргумент не 0, фукнція вертає результат ділення, інакше - замість виконання ділення, вертає повідомлення на екран "Не можу виконати ділення на 0".

    Якщо типи, надані у функцію, не `int` - функція виводить повідомлення "Не працюю з такими типами даних.".

  2. Запустіть функцію на наступних парах чисел:
- 12 і 6
- 20 і 0
- 112 і 10
- 99 і 0
- 11 i 1.1

    і виведіть на екран повідомлення "Результат роботи функції: " - і далі те, що повернула функція.

  3. Дайте відповідь на питання: "Чому в деяких випадках функція повернула None?"

````python
def int_division(a, b):
    if not isinstance(a, int) or not isinstance(b, int):
        print("Не працюю з такими типами даних.")
        return
    if b == 0:
        print("Не можу виконати ділення на 0")
        return
    return a / b
````

````python
tests = [(12, 6), (20, 0), (112, 10), (99, 0), (11, 1.1)]

for a, b in tests:
    result = int_division(a, b)
    print("Результат роботи функції:", result)
````

Функція повернула None, тому що в цих випадках вона не виконувала return з результатом. Функція нічого не повернула, лише виконала перевірку і повідомлення
