````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ["apple", "banana", "cherry", "dates", "elderberry"]
categories = ["fruit", "fruit", "fruit", "fruit", "fruit"]

N_ROWS = 10 ** 6

df = pd.DataFrame({
    "product_name": np.random.choice(products, N_ROWS),
    "product_category": np.random.choice(categories, N_ROWS),
    "product_price": np.random.uniform(1, 10, N_ROWS),
    "transaction_date": pd.date_range(
        start="1/1/2000",
        periods=N_ROWS,
        freq="h"
    )
})

# Вимірювання використання пам'яті до оптимізації
memory_before = df.memory_usage(deep=True).sum()

# Перетворення текстових колонок у категорійний тип
df["product_name"] = df["product_name"].astype("category")
df["product_category"] = df["product_category"].astype("category")

# Вимірювання використання пам'яті після оптимізації
memory_after = df.memory_usage(deep=True).sum()

# Обчислення коефіцієнта зменшення використання пам'яті
reduction_factor = memory_before / memory_after

print(f"Memory before: {memory_before / 1024 ** 2:.2f} MB")
print(f"Memory after: {memory_after / 1024 ** 2:.2f} MB")
print(f"Memory usage reduced by {reduction_factor:.2f} times")
````

Цей код:
- перетворює текстові колонки у категорійний тип;
- зменшує використання пам’яті;
- показує використання пам’яті до та після оптимізації.

````python
import warnings

import pandas as pd
import sweetviz as sv

# Завантаження даних
df = pd.read_csv(
    "ecommerce_data.csv",
    encoding="ISO-8859-1"
)

# Заповнення пропущених значень та зміна типу колонки
df["CustomerID"] = (
    df["CustomerID"]
    .fillna(-1)
    .astype(str)
)

# Фільтрація даних по країнах
germany_df = df[df["Country"] == "Germany"]
france_df = df[df["Country"] == "France"]

# Вимкнення попереджень та створення звіту
with warnings.catch_warnings(record=True):

    report = sv.compare(
        [germany_df, "Germany"],
        [france_df, "France"]
    )

    report.show_html("germany_vs_france_report.html")
````

````python
def calculate_variance(numbers):
    """
    Calculate the variance of a list of numbers.

    Args:
        numbers (list): A list of numeric values.

    Returns:
        float: The variance of the input numbers.
    """

    mean = sum(numbers) / len(numbers)

    variance = sum(
        (number - mean) ** 2
        for number in numbers
    ) / len(numbers)

    return variance


input_list = [75, 80, 85, 90, 95]

result = calculate_variance(input_list)

print(result)
````
