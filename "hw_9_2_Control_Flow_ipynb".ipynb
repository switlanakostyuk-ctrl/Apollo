1. **Перевірка температури:**
   - Завдання:
     - Дано заздалегідь визначену змінну температури `temperature`. Перевірте її значення.
     - Якщо температура вище 25 градусів Цельсія, виведіть "Це теплий день."
     - Якщо температура знаходиться в межах від 15 до 25 градусів Цельсія (включно), виведіть "Це приємний день."
     - В іншому випадку виведіть "Це холодний день."

```python
temperature = 20
```

```python
import random

temperature = 25

if temperature > 25:
    print("Це теплий день.")

elif 15 <= temperature <= 25:
    print("Це приємний день.")

elif 1 <= temperature < 15:
    print("Це холодний день.")

elif -10 <= temperature <= 0:
    frost_messages = [
        "Сьогодні морозно",  "На вулиці добрячий мороз",  "Сьогодні справжня холоднеча"
    ]
    print(random.choice(frost_messages))

elif -20 <= temperature < -10:
    strong_frost_messages = [
        "На вулиці тріскучий мороз",   "Сьогодні щипає за ніс"
    ]
    print(random.choice(strong_frost_messages))

elif -30 <= temperature < -20:
    print("На вулиці дубак.")

else:
    print("Екстремальні погодні умови.")
```

     - Якщо оцінка 90 або вище, виведіть "A".

```python
score = 75

if score >= 90:
    print("A")
elif 80 <= score <= 89:
    print("B")
elif 70 <= score <= 79:
    print("C")
elif 60 <= score <= 69:
    print("D")
else:
    print("F")
```
